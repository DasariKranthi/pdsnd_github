```r
#Data Wrangling - We want to combine all these three datasets to make it easier to run our queries.
#Adding Gender and Birth Year columns the Washington dataset before concatenating the datasets and set the values to NULL. It makes the concatination easy.
wash$Gender <- NA
wash$Birth.Year <- NA

#Adding a new column 'City' to each dataset to make it easy retain information about city after concatenation.
wash$City <- 'Washington'
ny$City <- 'New York'
chi$City <- 'Chicago'
```

### Question 1


**What are the counts of each gender (only available for NYC and Chicago)?**

```r
library(ggplot2)
```

```r
#Here omitting Washington data because it doesn't have Gender information.
#Created new dataframe by binding 'New York City' and 'Chicago' data
#Here I used tail to view the data, so we can also see the data where Gender is blank
df <- rbind(chi,ny)
tail(df)
```

```r
#From above summary we can see the data where Gender is not specified.
#Created another dataframe by omitting the NA and blank values from Gender column
#Used tail to view the dataframe so its easy to compare the data with first dataframe(df).
df2 = subset(df, !(is.na(df$Gender) | df$Gender==""))
tail(df2)
```

```r
# Using the Original dataset 'df' with non gender specified users
# Count of Gender (Male, Female and Not specified)
Total = sort(table(df$Gender))
print(Total)

# percentage of Gender (Male, Female and Not specified)
round((Total / length(df$Gender) * 100), digits = 2)
```

```r
# Data Visualization with ggplot using the 'df2' dataset (without non gender specified users)
ggplot(aes(x = Gender, fill = City), data = df2) +
    geom_bar(position = 'dodge') +
    ggtitle('Counts of each gender (only available for NYC and Chicago)') +
    labs(y = 'Number of Riders', x = 'Gender')
```

```r
# Count and Percentage of Gender(Male and Female) by city with non gender specified users
"Chicago data"
Chicago.Total = sort(table(df$Gender[df$City == 'Chicago']))
Chicago.Total
round((Chicago.Total / length(df$Gender[df$City == 'Chicago']) * 100), digits = 2)

"New York Data"
Newyork.Total = sort(table(df$Gender[df$City == 'New York']))
Newyork.Total
round((Newyork.Total / length(df$Gender[df$City == 'New York']) * 100), digits = 2)
```

```r
# Data Visualization with ggplot using the oroginal 'df' dataset (with non gender specified users)
ggplot(aes(x = Gender, fill = City), data = df) +
    geom_bar(position = 'dodge') +
    ggtitle('Counts of each gender (only available for NYC and Chicago)') +
    scale_x_discrete(labels = c('Not Specified', 'Female', 'Male')) +
    labs(y = 'Number of Riders', x = 'Gender')
```

Here, we have ommitted Washington data as it doesn't have the Gender information.

In Chicago City, there are 59.78% of Male (a total of 5159) , 19.97% of Female (a total of 1723) and 20.25% of Not Mentioned (a total of 1748).
In New York City, we have 67.92% of Male (a total of 37201) , 22.20% of Female (a total of 12159) and 9.88% of Not Mentioned (a total of 5410).

### Question 2

**What is the average travel time for users in different cities?**

```r
library(ggplot2)
```

```r
#Count of users in each city
Total = sort(table(bikeshare$City))
print(Total)

#percentage of users in each city
round((Total / sum(Total) * 100), digits = 2)
```

```r
#Calculating the Average travel time(AvgTT)

AvgTT <- with(bikeshare, aggregate(list(Trip.Duration), by = list(City), 
                   FUN = function(x) { mon.mean = mean(x, na.rm = TRUE) } ))

colnames(AvgTT) <- c('City', 'Average Trip Duration')
AvgTT
```

```r
#Data Visualization with ggplot
ggplot(aes(x = City, y = Trip.Duration), data = bikeshare) +
    geom_bar(fill = 'pink', stat = "summary", fun.y = "mean") + 
    ggtitle('The Average Travel Time for users in different cities') +
    labs(y = 'Travel Time', x = 'City')
```

**Summary of your question 2 results goes here.**

More that half of the Trip Duration is from Washington(1233.9533) due to more number of users.
The average trip duration in Chicago(937.1728) and New York City(903.6147) cities is almost the same.
Although the number of users in Chicago is less, average trip duration is almost equal with New York City.
Percentage count of users in Chicago is 5.66%, New York City is 35.93% and Washington is 58.41%.

### Question 3

**What is the most common month?**
**What is the most common day of week?**

```r
# Creating new columns 'Month' and 'Day.of.Week' extracting from Start.Time and change the data type to numeric
bikeshare$Month <- as.numeric(substring(bikeshare$Start.Time,6,7))
bikeshare$Day.of.Week <- (factor(weekdays(as.Date(bikeshare$Start.Time)), 
                 levels = c('Monday', 'Tuesday', 'Wednesday',
                            'Thursday', 'Friday', 'Saturday', 'Sunday')))
head(bikeshare)
```

```r
# Visualizing data with ggplot using 'Month'
ggplot(data = bikeshare, aes(x = Month)) +
    geom_histogram(stat="count", color='black') +
    facet_wrap(~City, scales = "free") +
    ggtitle('Monthly Trips by City') +
    ylab('Trip count')
```

```r
# Visualizing data with ggplot using 'Day.of.Week'
ggplot(data = bikeshare, aes(x = Day.of.Week)) +
    geom_histogram(stat="count", color='black') +
    facet_wrap(~City, scales = "free") +
    theme(axis.text.x = element_text(angle = 90)) +
    ggtitle('Weekly Trips by City') +
    ylab('Trip count')
```

**Summary of your question 3 results goes here.**

On monthly trip, June is the most used month in all three cities. Number of Trips in the month June is almost equal in all cities.
An explanation to this would be that the bikeshare service is more used in the spring months than on winter. 
It is interesting to note that the difference between the use in winter and spring months is stronger in Chicago and New York than in Washington.

On weekly trip, we see a higher use of the service on weekdays. This indicates that the service is quite used for work commutes in these cities. 
At Washington the service is lower on the weekends, but still at the level of Monday use.
In Chicago, the usage level is almost the same on all weekdays compared to New York and Washington cities where the highest usage is on a Wednesday.
